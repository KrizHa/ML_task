1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Створюємо значення x у діапазоні [0, 10]
x = np.linspace(0, 10)
f_lin = 10 * x - 5

# Oбчислення похідної

df_dx = np.diff(f_lin) / np.diff(x)

print("Похідна функції:", df_dx[0])

#Побудова графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_lin, label='f(x) = 10x-5', color='blue')
plt.plot(x[:-1], df_dx, label="f'(x)", color='red', linestyle='dashed')
plt.title("Графік функції 10x-5 та її похідної")
plt.xlabel("Кількість годин, витрачених на навчання")
plt.ylabel("Оцінка")
plt.grid(True)
plt.legend();
```

Кількість годин навчання лінійно і стрімко впливає на оцінку, чим більше годин тим вица оцінка


2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
from sympy.utilities.lambdify import lambdify

# Визначаємо символічну змінну
x = sp.symbols('x')

#Функція сінуса та іі похідна
f_sin = sp.sin(x)
df_sin = sp.diff(f_sin, x)

#Функція логарифму та іі похідна
f_ln = sp.ln(x)
df_ln = sp.diff(f_ln, x)

# Створюємо значення x_vals у діапазоні [0.5, 10]
x_vals = np.linspace(0.5, 10, 100)

#Перетворення символічних функцій в числові
f_sin_numeric, df_sin_numeric, f_ln_numeric, df_ln_numeric = map(lambda f: lambdify(x, f, 'numpy'), [f_sin, df_sin, f_ln, df_ln])

#Обчислюємо функцію сінуса та похідну зі значення з x_vals
y_sin = f_sin_numeric(x_vals)
y_df_sin = df_sin_numeric(x_vals)

#Обчислюємо функцію логаріфму та похідну зі значення з x_vals
y_ln = f_ln_numeric(x_vals)
y_df_ln = df_ln_numeric(x_vals)

# Вивід значень похідних
print("\n Похідна функції sin(x):", df_sin)
print("\n Похідна функції ln(x):", df_ln)

# Побудова графіка

fig, ax = plt.subplots(2, figsize=(16, 10))

ax[0].plot(x_vals, y_sin, label="f(x) = sin(x)", color="blue")
ax[0].plot(x_vals, y_df_sin, label="f'(x) = cos(x)", color="red", linestyle="dashed")


ax[1].plot(x_vals, y_ln, label="f(x) = ln(x)", color="black")
ax[1].plot(x_vals, y_df_ln, label="f'(x) = 1/x", color="green", linestyle="dashed")


ax[0].set_xlabel("x")
ax[1].set_xlabel("x")
ax[0].set_title("Функція sin(x) та її похідна")
ax[1].set_title("Функція ln(x) та її похідна")
ax[0].legend()
ax[1].legend()

plt.show();
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

#Функція витрат виробництва та іі похідна
f_expenses = x**2 - 5*x + 1
df_expenses = sp.diff(f_expenses, x)


print("\n Похідна функції:", df_expenses)

# Створюємо значення x_vals у діапазоні [0, 100]
xs = np.linspace(0, 100, 100)

#Перетворення символічних функцій в числові

f_expenses_np, df_expenses_np = map(lambda f: lambdify(x, f, 'numpy'), [f_expenses, df_expenses])

#Обчислюємо функцію та похідну зі значеннями xs
ys = f_expenses_np(xs)
dydx = df_expenses_np(xs)

# Побудова графіка
plt.figure(figsize=(8, 4))
plt.plot(xs, ys, label="f(x):x^2 - 5x + 1", color="blue")
plt.plot(xs, dydx, label=f"f'(x) = {df_expenses}", color="red", linestyle="dashed")

plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Витрати компанії")
plt.title("Загальні витрати компанії від обсягу виробництва")
plt.legend()
plt.grid(True)
plt.show();
```
